#### 3. Задание (в программе или на листочке)
На листе тетради «в линейку» (расстояние между линиями равно а) лежит игла (длиной b). Координаты нижней точки иглы (х,у), игла лежит под углом alfa. Пересекает ли игла линию или нет?


```python
%matplotlib inline
import numpy as np
from matplotlib import pyplot as plt
import matplotlib.ticker
import math
```

```python
fig= plt.figure(figsize=(10,10))

##-------линии---------

## пусть betta - угол наклона линий
betta=65*np.pi/180

## если a - расстояние между линии, то отрезки междк ними вдоль оси Y будут равны a/cos(betta)
a=2

x=np.linspace(-10,10,1000)

## цикл по начертпнию линий
for i in range(-10,11):
    plt.plot(x, np.tan(betta)*x+i*a/np.cos(betta), color='grey')

#-----игла--------------
b=2
alpha=60*np.pi/180

# нижняя точка иглы
(x1, y1)=(-5,-5)

# тогда ккординаты конца:

(x2, y2)=(x1+b*np.cos(alpha), y1+b*np.sin(alpha))

#начертим иглу, при помощи уравнения прямой, проходящей черех 2 точки
if (alpha<=np.pi/2) or (alpha>=3*np.pi/2):
    x=np.linspace(x1,x2,100)
else:
    x=np.linspace(x2,x1,100)
    
plt.plot(x, y1+(x-x1)*(y2-y1)/(x2-x1), color='red')


plt.ylim(-20,20)
```

видно, что игла не пересекает линию.

Решим задачу численным методом:

```python
from scipy.optimize import fsolve

for i in range(-10,11):
    b1=i*a/np.cos(betta)
    def eq(p, b1=b1):
        x,y=p
        return (y-np.tan(betta)*x-b1,y-y1-(x-x1)*(y2-y1)/(x2-x1))
    x_per,y_per=fsolve(eq,(i,0))
    if (((alpha<=np.pi/2) and (x_per>=x1) and (x_per<=x2) and (y_per>=y1) and (y_per<=y2)) or
    ((alpha<=np.pi) and (x_per<=x1) and (x_per>=x2) and (y_per>=y1) and (y_per<=y2)) or
    ((alpha<=3*np.pi/2) and (x_per>=x1) and (x_per<=x2) and (y_per<=y1) and (y_per>=y2))or
    ((alpha<=2*np.pi) and (x_per>=x1) and (x_per<=x2) and (y_per<=y1) and (y_per>=y2))):
        print('с линией kx+', i, ' точка пересечения',np.round(x_per,2),np.round(y_per,2))
    else:
        print('нет пересечения с линией kx+', i)
    
    
    
```
