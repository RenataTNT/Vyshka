### 4. Задание (в программе)
Решите систему уравнений:

y = x2 – 1

exp(x) + x∙(1 – y) = 1


```python
from scipy.optimize import fsolve

def eq(p):
    x,y=p
    return (y-x**2+1, np.exp(x)+x*(1-y)-1)

for i in range(-5,5):
    x1,y1=fsolve(eq,(i,i**2-1))
    print(x1,y1)

```

видимо, из-за того, что в 0 2-е уравнение не определено из-за деления на 0, метод в качестве еще одного корня выдал значение 1-го уранения в 0-й точке (0,-1)

```python
fig= plt.figure(figsize=(5,5))
x=np.linspace(-10,10,1000)
plt.plot(x,x**2-1)
plt.plot(x,(np.exp(x)+x-1)/x)
plt.xlim(-10,10)
plt.ylim(-5,50)

plt.grid()
```
